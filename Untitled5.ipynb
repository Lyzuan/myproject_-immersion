```python
import requests
from bs4 import BeautifulSoup
import re

# Função para cadastrar o cliente no aplicativo SA Esportes (simulado)
def cadastrar_cliente(nome):
    # Simulação de cadastro - em um sistema real, integraria com a API do SA Esportes
    print(f"Cadastrando {nome} no SA Esportes...")
    # Em um sistema real, você enviaria uma solicitação para a API do SA Esportes
    # para cadastrar o cliente.
    print(f"{nome} cadastrado com sucesso!")

# Função para validar o código do bilhete no aplicativo SA Esportes (simulado)
def validar_codigo_bilhete(codigo):
    # Simulação de validação - em um sistema real, integraria com a API do SA Esportes
    print(f"Validando código do bilhete: {codigo}...")
    # Em um sistema real, você consultaria a API do SA Esportes para validar o código.
    if len(codigo) == 6 and codigo.isdigit():
        return True
    else:
        return False

# Função para gerar o link de pagamento via Pix (simulado)
def gerar_link_pagamento(valor):
    # Simulação de geração de link - em um sistema real, integraria com a plataforma de pagamento
    print(f"Gerando link de pagamento via Pix para R${valor:.2f}...")
    # Em um sistema real, você utilizaria a API da plataforma de pagamento para gerar o link.
    return "https://link_de_pagamento_simulado.com"

# Função para gerar o link do bilhete no SA Esportes (simulado)
def gerar_link_bilhete(codigo):
    # Simulação de geração de link - em um sistema real, integraria com a API do SA Esportes
    print(f"Gerando link do bilhete...")
    # Em um sistema real, você usaria a API do SA Esportes para gerar o link do bilhete.
    return f"https://link_do_bilhete_simulado.com/{codigo}" # Inclui o código no link

# Função para consultar e enviar a segunda via do bilhete (SIMULADO)
def conferir_bilhete(codigo):
    # Simula consulta em um banco de dados ou API - adicione códigos válidos aqui
    bilhetes_validos = {
        "ABCD1234": {"data": "2023-12-28", "valor": 50.00, "status": "Aberto"},
        "EFGH5678": {"data": "2023-12-27", "valor": 100.00, "status": "Ganhador"},
        "IJKL9012": {"data": "2023-12-26", "valor": 25.00, "status": "Perdedor"}
    }

    if codigo in bilhetes_validos:
        bilhete = bilhetes_validos[codigo]
        print(f"Informações do bilhete {codigo}:\n"
              f"Data: {bilhete['data']}\n"
              f"Valor: R${bilhete['valor']:.2f}\n"
              f"Status: {bilhete['status']}\n")
        print(f"[Link simulado da segunda via do bilhete: https://link_do_bilhete_simulado.com/{codigo}]")
        return True # Retorna True se o bilhete for encontrado
    else:
        print("Código inválido. Por favor, tente novamente.")
        return False # Retorna False se o bilhete não for encontrado

# Loop principal do chatbot
while True:
    print("\nOlá, sou o Max, seu assistente de apostas!")
    print("1 - Realizar aposta")
    print("2 - Conferir bilhete")
    print("3 - Finalizar atendimento")

    opcao = input("Digite o número da opção desejada: ")

    if opcao == "1":
        nome = input("Você já possui cadastro? (s/n): ")
        if nome.lower() == 's':
            nome = input("Digite seu nome cadastrado: ")
        else:
            nome = input("Digite seu nome para cadastro: ")
            cadastrar_cliente(nome)

        while True:
            valor_aposta = float(input("Digite o valor da aposta (limite de R$1000): R$"))
            if valor_aposta > 1000:
                print("Valor da aposta excede o limite de R$1000. Tente novamente.")
            else:
                link_pagamento = gerar_link_pagamento(valor_aposta)
                print(f"Link para pagamento via Pix: {link_pagamento}")
                break  # Sai do loop de valor da aposta

        print("Para concluir sua aposta, insira o código de 6 dígitos numéricos.")
        while True:  # Loop para solicitar o código até que seja válido
            codigo_bilhete = input("Digite o código do bilhete (6 dígitos numéricos): ")
            if validar_codigo_bilhete(codigo_bilhete):
                link_bilhete = gerar_link_bilhete(codigo_bilhete)
                print(f"Bilhete gerado com sucesso! O código do seu bilhete é: {codigo_bilhete}. Você pode acessá-lo aqui: {link_bilhete}")
                print("Boa sorte!")
                print("Agradecemos por utilizar nossos serviços. Até a próxima!")
                break  # Sai do loop de código do bilhete
            else:
                print("Código inválido. Por favor, tente novamente.")
        break  # Encerra o atendimento após finalizar a aposta

    elif opcao == "2":
        print("Para conferir seu bilhete, você precisa do código de 8 dígitos (letras maiúsculas e números) que foi fornecido ao criar a aposta.")
        while True:
            codigo_bilhete = input(
                "Digite o código do bilhete (8 dígitos, letras maiúsculas e números): "
            )
            bilhete_encontrado = conferir_bilhete(codigo_bilhete)
            if bilhete_encontrado:
                break  # Sai do loop se o bilhete for encontrado

    elif opcao == "3":
        print("Agradecemos por utilizar nossos serviços. Até a próxima!")
        break
    else:
        print("Opção inválida. Por favor, digite um número entre 1 e 3.")
```
